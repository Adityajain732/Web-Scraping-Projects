```python
import string
string.punctuation
```

```python
def text_process(mess):
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    words = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    return [item for item in words if item.isalpha() and item.lower()!='kg' and item.lower()!='kgs']
```

```python
data['Product description'].head()
```

```python
data['Product description'].head(5).apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
bow_transformer = CountVectorizer(analyzer=text_process).fit(data['Product description'])
```

```python
pro4 = data['Product description'][3]
print(pro4)
```

```python
bow4 = bow_transformer.transform([pro4])
print(bow4)
print(bow4.shape)
```

```python
print(bow_transformer.get_feature_names())
```

```python
product_bow = bow_transformer.transform(data['Product description'])
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer().fit(product_bow)
tfidf4 = tfidf_transformer.transform(bow4)
print(tfidf4)
print(tfidf4.shape) 
```
